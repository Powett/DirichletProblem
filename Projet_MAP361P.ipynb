## I) En dimension 1

\begin{equation*}
\tag{1.1}
\begin{cases}
u''(x)=f(x)\ \forall x \in I,\\

\label{D1}

T1) Si u et v sont des solutions de \eqref{D1}, alors $g=u-v$ est solution de (1) pour $f=0,\ a=0$ et $b=0$.

  Alors $g=\alpha x+\beta$.

  Mais $g(0)=g(1)=0$, donc $\alpha=\beta=0$, finalement $u=v$.

  Donc \eqref{D1} admet au plus une solution.

  Si $f(x)=\sin(2\pi x), a=b=0$, on obtient : 
  $$
  u(x)=-\frac{1}{(2\pi)^2}\sin(2\pi x)
  $$

  En effet en dérivant deux fois on obtient bien $f$ et $u(0)=u(1)=0$.

  Si $f=0$, $a$ et $b$ quelconques, on obtient $u(x)=(b-a)x+a$.

T2) Si $u\in C^4(I)$, on a si $h\to 0$ :

\begin{cases}
u(x+h)&=u(x)+u'(x)h+u"(x)\frac{h^2}{2}+u^{(3)}(x)\frac{h^3}{6}+O(h^4)\\
u(x-h)&=u(x)-u'(x)h+u"(x)\frac{h^2}{2}-u^{(3)}(x)\frac{h^3}{6}+O(h^4)\\
\end{cases}
\end{equation*}
Ainsi en sommant les deux expressions
$$
\frac{u(x-h)-2u(x)+u(x+h)}{h^2}=u"(x)+O(h^2)

T^x=inf\{T^{x}_0,T^{x}_1\}
\end{cases}

$\mathbb{P}(T^x>2M|T^x>M)$ correspond à la probabilité que le point étudié sorte de l'intervalle \]0,1\[ à un instant strictement supérieur à 2M, en sachant qu'il n'en était pas sorti avant M (compris).\
A tout instant, si le point se déplace M fois d'affilée d'un côté, il est certain qu'il sortira de I, car il parcourra une distance supérieure ou égale en valeur absolue à 1.\
Formellement, on a :
\begin{equation*}
\{T^x\leq2M|T^x>M\}\supset\{\zeta_{M+1}=1,\zeta_{M+2}=1,....\zeta_{2M}=1\}\cup\{\zeta_{M+1}=-1,\zeta_{M+2}=-1,....\zeta_{2M}=-1\}
\end{equation*}
Les variables aléatoires $(\zeta_{i})_{M+1\leq i\leq 2M}$ sont indépendantes de $(\zeta_{i})_{0\leq i\leq M}$,\
Les deux éléments indiqués à droite sont disjoints et de probabilité identique, égale à $\frac{1}{2^{M}}$.\
Ainsi on retrouve 
$$
\mathbb{P}(T^x\leq2M|T^x>M)\geq \frac{1}{2^{M-1}}\geq \frac{1}{2^{M}}
$$
Finalement,
$$
\mathbb{P}(T^x>2M|T^x>M)=1-\mathbb{P}(T^x\leq2M|T^x>M)\leq 1- \frac{1}{2^M}
$$
Ce raisonnement est en fait valable pour tout $k\in\mathbb{N}$ :
$$
\forall k\in\mathbb{N}, \mathbb{P}(T^x>(k+1)M|T^x>kM)\leq 1- \frac{1}{2^M}
$$
De plus pour $k=0$, on a $\mathbb{P}(T^x>M|T^x>0)=\mathbb{P}(T^x>M)$ car $T^x$ est nécessairemment strictement positif ($x\in I=\ ]0,1[ $).
De plus par définition, 
$$
\mathbb{P}(T^x>(k+1)M|T^x>kM)=\frac{\mathbb{P}(\ (T^x>(k+1)M)\cap (T^x>kM)\ )}{\mathbb{P}(\ T^x>kM\ )}=\frac{\mathbb{P}(T^x>(k+1)M)}{\mathbb{P}(T^x>kM)}
$$
On obtient finalement $\forall k\in \mathbb{N}^*$:
\begin{equation*}
\prod^{k-1}_{i=0} \frac{\mathbb{P}(T^x>(i+1)M)}{\mathbb{P}(T^x>iM)}=\frac{\mathbb{P}(T^x>kM)}{\mathbb{P}(T^x>0)}=\mathbb{P}(T^x>kM)
\end{equation*}\
\
\
D'où $\forall k\in \mathbb{N}^*$ :

\begin{equation*}
a_k=\mathbb{P}(T^x>kM)\leq(1-\frac{1}{2^M})^k
\end{equation*}

M étant un entier strictement positif, la famille $(p_k)_{k\geq 1}$ est sommable (somme géométrique de raison <1).
\
Soit à présent une variable aléatoire Y à valeurs dans $\mathbb{N}^*$ quelconque, telle que $(p_k=\mathbb{P}(Y\geq k))_{k\geq 1}$ soit sommable.\

\begin{equation*}
\sum_{n=1}^\infty p_n=\sum_{n=1}^\infty \sum_{k=n}^\infty \mathbb{P}(Y=k)
\end{equation*}

En effet, $\{Y\geq n\}=\cup_{k\geq n}\{Y=k\}$, et cette union est disjointe.
Ainsi on obtient après permutation des sommes (théorèmes de sommation par paquets et de Fubini) :

\begin{equation*}
\sum_{n=1}^\infty p_n=\sum_{k=1}^\infty k\mathbb{P}(Y=k)=\mathbb{E}(Y)
\end{equation*}

On a ainsi montré que Y est intégrable (tous les termes présents sont positifs) et que $\mathbb{E}(Y)=\sum_{n=1}^\infty p_n$.
Appliqué à $\frac{T^x}{M}$, on obtient comme $(a_k=\mathbb{P}(\frac{T^x}{M}\geq k))_{k\geq 1}$ est sommable,

\begin{equation*}
\mathbb{E}(\frac{T^x}{M})=\sum_{n=1}^\infty a_n <\infty
\end{equation*}
\
En particulier, par linéarité de l'espérance on obtient que $\mathbb{E}(T^x)=M\mathbb{E}(\frac{T^x}{M})<\infty$.

```python

### Valeurs génériques ###

x=1/2
h=5e-2
######

#S1)
def generate_traj(x,h): #fonction générant aléatoirement une trajectoire, jusqu'à la sortie de ]0,1[
  X = [x]
  while x>0 and x<1:
    k = npr.choice([1,-1])
    x+= k*h
    X.append(x)
  X.append(x)
  return X

def generate_end(x,h): #fonction similaire qui permet de récupérer directement l'indice et la position de fin (utile pour gagner du temps par la suite)
  n=0
  while x>0 and x<1:
    k = npr.randint(0,2)*2-1
    x+= k*h
    n+=1
  return x,n

def s1():
  for i in range(5):
    traj=generate_traj(x,h)
    n=len(traj)
    plt.plot(np.arange(0,n,1), traj)
  plt.legend(["Trajectoire "+ str(i+1) for i in range(5)])
  plt.title(f"Trajectoires, h={h}, x={x}")
  plt.show()
```

Et de manière tout à fait analogue, $$\mathbb{P}( X_{T^x} = 1 ) = \frac{1}{2}\mathbb{P}( X_{T^{x+h}} = 1 ) + \frac{1}{2}\mathbb{P}( X_{T^{x-h}} = 1 ) $$

De plus, par linéarité de l'espérance, $v_h(x) = a \mathbb{P}(X_{T^x} = 0) + b\mathbb{P}(X_{T^x} = 1) $
\begin{align*}
 \Delta_hv_h(x) &= \frac{v_h(x+h) -2v_h(x) + v_h(x-h)}{h^2}\\
 &=  \frac{v_h(x+h) + v_h(x-h) -2 \left[a(\frac{1}{2}\mathbb{P}( X_{T^{x+h}} = 0 ) + \frac{1}{2}\mathbb{P}( X_{T^{x-h}} = 0 )) + b(\frac{1}{2}\mathbb{P}( X_{T^{x+h}} = 1 ) + \frac{1}{2}\mathbb{P}( X_{T^{x-h}} = 1 ))\right]}{h^2}\\

Donc $v_h$ est bien solution pour $f=0$.


```python
#S2)
#a)
def v(h,x,a,b,N):
  s = 0 
  for i in range(N):
    y,n=generate_end(x,h)
    if y >= 1 :
      s+=1
  return a* (N-s)/N + b * s/N #Valeur empirique des espérances pour v_h

def s2_a(fonction):
  a,b = 1 ,2
  N= 1000
  h= 0.05
  M= int(1/h)
  X = np.arange(0,M+1,1)*h
  if fonction=="v":
    Y = [v(h,x,a,b,N)  for x in X]
    Z=(b-a)*X+a
  elif fonction=="w":
    Y = [w(h,x,a,b,N)  for x in X]
    Z= -1/(4*np.pi**2)*np.sin(2*np.pi*X)
  plt.plot(X,Y)
  plt.plot(X,Z) #valeur exacte de la solution
  plt.title(f"Comparaison {fonction}_h(N) et solution exacte correspondante u")
  plt.legend([f"{fonction}_h","u"])
  plt.show()
```

$\to$ Si $x\neq h$ et $x\neq 1-h$

 Par la formule des espérances emboîtées, $\mathbb{E}\left(\sum_{n=0}^{T^x -1}f(X_n)\right) = \mathbb{E}\left(\mathbb{E}(\sum_{n=0}^{T^x -1}f(X_n) |X_1 )\right)$ 

Or
\begin{align*}
 \mathbb{E}\left(\sum_{n=0}^{T^x -1}f(X_n) |X_1 = x+h \right) &= \mathbb{E}\left(\sum_{n=0}^{T^{x+h} -1}f(Y_n) \right) + f(x) & &\text{où $Y_0= x+h$ et les $Y_n$ sont définis de la même manière récurrente que les $X_n$}\\

 \mathbb{E}\left(\sum_{n=0}^{T^x -1}f(X_n) |X_1 = x-h \right) &= \mathbb{E}\left(\sum_{n=0}^{T^{x-h} -1}f(Y_n) \right) + f(x) & &\text{où $Y_0= x-h$ et les $Y_n$ sont définis de la même manière récurrente que les $X_n$}\\
\end{align*}

On a finalement :

\begin{align*}
  w_h(x) &= -\frac{h^2}{2}\left[ \mathbb{E}\left(\mathbb{E}(\sum_{n=0}^{T^x -1}f(X_n) |X_1 ) \right) \right] \\
  &= -\frac{h^2}{2}\left[ \mathbb{E}\left(\sum_{n=0}^{T^x -1}f(X_n) |X_1= x+ h \right)\mathbb{P}(X_1 = x+h) + \mathbb{E}\left(\sum_{n=0}^{T^x -1}f(X_n) |X_1= x - h \right)\mathbb{P}(X_1 = x-h)  \right]\\
  &= -\frac{h^2}{2}\left[ \frac{1}{2}\mathbb{E}\left(\sum_{n=0}^{T^{x+h} -1}f(Y_n) \right) + \frac{1}{2}f(x) + \frac{1}{2}\mathbb{E}\left(\sum_{n=0}^{T^{x-h} -1}f(Y_n) \right) + \frac{1}{2}f(x)  \right]\\
 &= \frac{1}{2}\left[ -\frac{h^2}{2}\mathbb{E}\left(\sum_{n=0}^{T^{x+h} -1}f(X_n) \right)  -\frac{h^2}{2}\mathbb{E}\left(\sum_{n=0}^{T^{x-h} -1}f(X_n) \right) \right]-\frac{h^2}{2}f(x)&\text{car les $Y$ et $X$ suivent la même loi à l'indice de départ près}\\

On a donc bien $\frac{w_h(x+h)+w_h(x-h)-2w_h(x)}{h^2} = f(x)$

$\to$ Si $x = h$
$$\mathbb{E}\left(\sum_{n=0}^{T^x -1}f(X_n) |X_1 = x-h = 0 \right) = \mathbb{E}(f(X_0)) = f(x) = -\frac{2}{h^2} w_h(x-h) + f(x)$$ 

Puisque $w_h(x-h) = w_h(0)= 0$

On retombe donc sur la même formule que dans le cas précédent, et on a donc bien $w_h$ solution en $x = h$

$\to$ Si $x = 1-h$

Puisque $w_h(x+h) = w_h(1)= 0$

On retombe donc sur la même formule que dans le cas précédent, et on a donc bien $w_h$ solution en $x = 1-h$

```python
#S3)
def f(x):
  return np.sin(2*np.pi*x)
def w(h,x,a,b,N):
  s = 0 
  for i in range(N):
    y = x
    t = 0
    sim=f(y)
    while y > 0 and y < 1 :
      k = npr.choice([1,-1])
      y += k*h
      t+=1
      sim+=f(y)
    s+=sim
  return s*(-h**2/2)/N #Valeur empirique des espérances pour w_h
```

T6) On a déjà montré que $$\mathbb{E}\left(\sum_{n=0}^{T^x -1}f(X_n) |X_1 = x+h \right) = \mathbb{E}\left(\sum_{n=0}^{T^{x+h} -1}f(Y_n) \right) + f(x)$$ Or $$ \mathbb{E}(T^x|X_1 = x+h ) = \mathbb{E}\left(\sum_{n=0}^{T^x -1}f(X_n) |X_1 = x+h \right) $$ en posant $f$ la fonction constante égale à $1$.
D'où $$ \mathbb{E}(T^x|X_1 = x+h ) = \mathbb{E}\left(T^{x+h} \right)  + 1 \text{ et de même } \mathbb{E}(T^x|X_1 = x-h ) = \mathbb{E}\left(T^{x-h} \right)  + 1 $$
Par le même raisonnement que la question T5), la formule reste viraie aux cas limites $x = h$ et $x=1-h$.

On utilise donc la formule des espérances emboîtées : 
\begin{align*}
    \mathbb{E}(T^x) &= \mathbb{E}\left(\mathbb{E}\left({T^x }|X_1  \right) \right)\\
    &= \mathbb{E}\left({T^x }|X_1 =x+h \right) \mathbb{P}(X_1 =x+h) + \mathbb{E}\left({T^x }|X_1 =x-h \right) \mathbb{P}(X_1 =x-h) \\
    &= \frac{1}{2} (\mathbb{E}\left(T^{x+h} \right) +1) + \frac{1}{2} (\mathbb{E}\left(T^{x-h} \right) +1)\\
    &= 1+    \frac{\mathbb{E}\left(T^{x+h} \right) +\mathbb{E}\left(T^{x-h} \right)}{2} 
\end{align*}
Comme $x \in I_h$, on peut poser $ x = nh $ où $n\in [\![1; \frac{1}{h} -1 ]\!]$. On pose alors $\forall n \in [\![1; \frac{1}{h} -1 ]\!]$, $u_n = \mathbb{E}(T^{nh})$ et on a 
\begin{align*}
    2u_n &= 2 + u_{n+1} + u_{n-1} \\
    0 &= u_{n+1} - u_n + u_{n-1} - u_n +2 \\
    -2 &= v_{n+1} - v_n & &\text{où on a posé $\forall n \in [\![2; \frac{1}{h} -1 ]\!]$, $ v_n = u_n - u_{n-1}$}
\end{align*}

$(v_n)$ est donc une simple suite arithmétique, donc $\forall n \in [\![2; \frac{1}{h} -1 ]\!]$, $ v_n = -2n + c$ avec $c \in \mathbb{R}$

Ainsi $\forall n \in [\![2; \frac{1}{h} -1 ]\!]$
$$ u_n = u_{n-1} -2n +c $$
D'où $\forall n \in [\![1; \frac{1}{h} -1 ]\!]$
$$ u_n = \sum_{i=0}^n( -2n +c) + d \text{ avec } d \in \mathbb{R}$$
$$ u_n = -2 \frac{n(n+1)}{2} + n*c + d$$
$$ u_n = -n(n+1) + n*c + d$$

Les conditions aux limites imposent $u_0 = u_{\frac{1}{h}} = 0$ puisque $\mathbb{E}(T^0) = \mathbb{E}(T^1 )= 0$

Donc $d = 0 $ et $c= 1 + \frac{1}{h}$, et 
$$ u_n = n(\frac{1}{h} -n )$$

Finalement, 
\begin{align*}
  \mathbb{E}(T^{nh}) &= n(\frac{1}{h} -n )\\
  \mathbb{E}(T^{x}) &= \frac{x}{h}(\frac{1}{h} -\frac{x}{h} )\\
  &= \frac{x(1-x)}{h^2}
\end{align*}

On retrouve bien le résultat numérique précédent avec $\mathbb{E}(T^{x}) = c*h^\alpha$ avec $\alpha \simeq -2$


T7) On remarque que le raisonnement effectué précédemement est toujours valable : quel que soit le point $X_n$, se déplacer M fois dans une direction donnée (il en existe à présent 4 au lieu de 2) fait nécessairement sortir de $D_h$ : 
\begin{equation*}
\{T^{x,y}_{\partial D_h} \leq2M|T^{x,y}_{\partial D_h} >M\}\supset\bigcup_{i\in(0,1),(1,0),(0,-1),(-1,0)} \{\eta_{M+1}=i,\eta_{M+2}=i,....\eta_{2M}=i\}
\end{equation*}
Tous ces évènements sont disjoints et de probabilité $\frac{1}{4^M}$, ainsi on obtient :
\begin{equation*}
\mathbb{P}(T^{x,y}_{\partial D_h}>2M|T^{x,y}_{\partial D_h}>M)=1-\mathbb{P}(T^x\leq2M|T^x>M)\leq 1- \frac{1}{4^M}
\end{equation*}
On obtient finalement, après les mêmes calculs,
$$
\mathbb{E}(T^{x,y}_{\partial D_h})<+\infty
$$

T8)
Soit $\forall (x,y)\in D_h\ ,u_h(x,y)=\mathbb{E}(g(X_{T^{x,y}_{\partial D_h}}))$

- Si $(x,y)\in \partial D_h$, il est assez clair que $T^{x,y}_{\partial D_h}=0$, donc $u_h(x,y)=g(x,y)$.

- Si $(x,y)\in D\setminus \partial D_h$

  Comme précédemment, on peut conditionner cette espérance par $\eta_1$, puis à $\eta_1$, donc $X_1$ fixé, on peut décaler les indices de 1 : $\{X_{T^{x,y}_{\partial D_h}}|X_1=z\}$ suit la même loi que $\{X_{T^{z}_{\partial D_h}}\}$ (z dans D).

  On obtient donc:
  \begin{equation*}
  u_h(x,y)=\frac{1}{4}\left( u_h(x+h,y)+u_h(x-h,y)+u_h(x,y+h)+u_h(x,y-h)\right)
  \end{equation*}
  Donc on a exactement, sous la même condition :
  \begin{equation*}
  h^2\Delta_h u_h(x,y)= \left( u_h(x+h,y)+u_h(x-h,y)+u_h(x,y+h)+u_h(x,y-h)\right) -4u_h(x,y)=0
  \end{equation*}
Finalement on a :
\begin{cases}
\Delta_h u_h(x,y)&=0&\forall (x,y)\in D\\
u_h(x,y)&=g(x,y)&\forall (x,y)\in\partial D_h
\end{cases}
Ce qui signifie bien que $u_h$ est une solution de (1.6).
Comme $\Delta_h$ est une approximation de $\Delta$, et $D_h$ de D, il semble logique que **$u_h$ soit une solution approchée du problème de Dirichlet initial, avec f=0** :
\begin{cases}
\Delta u(x,y)&=0&\forall (x,y)\in D\\
u(x,y)&=g(x,y)&\forall (x,y)\in\partial D
\end{cases}

```python
#S5)

def g(x,y):
  return np.exp(np.pi*y)*np.sin(np.pi*x)

def generate_end_2D(x,y,h):
  while h<=x<=1-h and h<=y<=1-h :
    rd=npr.randint(0,4)
    if rd==0:
      x+=h
    if rd==1:
      x-=h
    if rd==2:
      y+=h
    if rd==3:
      y-=h
  return (x,y)

def approx_uh(x,y,h,N):
  s=0
  for i in range(N):
    a,b=generate_end_2D(x,y,h)
    s+=g(a,b)
  return s/N

def s5():
  N=200
  h=2e-2
  # Attention : diminuer h augmente très rapidement le temps de calcul
  # Pour N=200, h=2e-2, le calcul prend environ 10min

  fig = plt.figure(figsize=(20,20))
  fig2=plt.figure(figsize=(20,20))
  ax = fig.add_subplot(111, projection='3d')
  ax2 = fig2.add_subplot(111, projection='3d')
  X=np.arange(0,1,h)
  Y=np.arange(0,1,h)
  X,Y=np.meshgrid(X,Y)
  Z=np.zeros_like(X)
  Zth=np.zeros_like(X)
  diff=np.zeros_like(X)
  for i in range(len(X)):
    for j in range(len(Y)):
      Z[j,i]=approx_uh(i*h,j*h,h,N)
      Zth[j,i]=g(i*h,j*h)
      if Zth[j,i]!=0:
        diff[j,i]=np.abs((Z[j,i]-Zth[j,i]))/Zth[j,i]
  ax.plot_surface(X,Y,Z,cmap=cm.coolwarm, alpha=0.95, label="Simulation")
  ax.plot_wireframe(X,Y,Zth, color="black", label ="Solution exacte", alpha=0.75)
  ax.set_xlabel("X")
  ax.set_ylabel("Y")
  ax.text2D(0, 1, r"Tracé de $u_h$ et de g",size=40, transform=ax.transAxes)
  ax2.text2D(0, 1, r"Erreur relative $\frac{|u_h-g|}{g}$",size=40, transform=ax2.transAxes)
  ax2.plot_surface(X,Y,diff, cmap=cm.copper, label="Erreur relative")
  ax2.set_xlabel("X")
  ax2.set_ylabel("Y")
s5()
```

On remarque qu'en prenant une petite valeur de h (ici 2e-2), l'approximation obtenue est plutôt bonne : la surface obtenue (en dégradé du bleu au rouge) est très proche de la surface de la solution exacte (en maillage noir), et l'erreur relative maximale est globalement sous les 20% (en fait les pics observés sont sûrement dûs à des problèmes d'arrondis lors du calcul sur des très petits nombres flottants).
